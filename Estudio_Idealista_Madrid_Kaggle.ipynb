```python
# PRIMERA ESTIMACION:
# SEGUN EL INE (INSTITUTO NACIONAL DE ESTADISTICA), CON DATOS DE 2022, PARA UNA POBLACION DE MADRID DE 1.162.155 HABITANTES DE 
# ENTRE 20 AÑOS A 34 AÑOS QUE PODRÍA SER MUY PROBABLE QUE ESTÉN CON INTENCIÓN DE ADQUIRIR SU PRIMERA VIVIENDA O SEGUNDA 
# OBTENEMOS LA SIGUIENTE TASA:
# 16 VIVIENDAS PARA 1.162.155 HABITANTES EN EL CASO DE ENDEUDAMIENTO A 30 AÑOS
# 23 VIVIENDAS PARA LA MISMA POBLACION EN EL CASO DE ENDEUDAMIENTO A 40 AÑOS.

# ACCORDING TO THE INE (INSTITUTE NATIONAL STATISTICS), WITH 2022 DATA, FOR A POPULATION OF MADRID OF 1,162,155 INHABITANTS 
# AGED BETWEEN 20 AND 34 YEARS WHO COULD BE VERY LIKELY TO BE INTENDING TO PURCHASE THEIR FIRST OR SECOND HOME, WE OBTAIN THE 
# FOLLOWING RATE:
# 16 HOUSES FOR 1,162,155 INHABITANTS IN THE CASE OF A 30-YEAR LOAN
# 23 HOUSES FOR THE SAME POPULATION IN THE CASE OF A 40-YEAR LOAN.


poblEntre20y34 = 1162155
```

```python
# Datos
datosEnX = ['30', '40']

cantidadEnY = [dfCantMenos100K/poblEntre20y34, dfCantMenos220K/poblEntre20y34]

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Anchura de las barras
width = 0.35

# Graficar la cantidad de casas
rects1 = ax.bar(datosEnX, cantidadEnY, width, label='Casas Disponibles/Poblacion')

# Agregar valores en las barras
for rect in rects1:
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2., height, '{:.5f}'.format(height),
            ha='center', va='bottom')





# Etiquetas y título
ax.set_xlabel('Años de Endeudamiento')
ax.set_ylabel('Población')
ax.set_title('Casas Disp. / Personas en Edad de Hipoteca')
ax.legend()

# Mostrar la gráfica
plt.show()
```

```python
# VEMOS QUE SE PARECE A UNA GRÁFICA CHI CUADRADA

# Define the dataset

# x = np.array([1,3,5,7,8,9, 10, 15])
x = dfViviendas['sqft']

#y = np.array([10, 20, 30, 40, 50, 60, 70, 80])
y = dfViviendas['price']


def Pearson_correlation(X,Y):
    if len(X)==len(Y):
        Sum_xy = sum((X-X.mean())*(Y-Y.mean()))
        Sum_x_squared = sum((X-X.mean())**2)
        Sum_y_squared = sum((Y-Y.mean())**2)       
        corr = Sum_xy / np.sqrt(Sum_x_squared * Sum_y_squared)
    return corr
             
print("Correlación Pies Cuadrados/Precio: ",Pearson_correlation(x,y))    
print("Comprobación de correlación: ",Pearson_correlation(x,x))
```
